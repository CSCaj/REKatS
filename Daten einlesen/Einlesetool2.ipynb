```python
# Marquee Fortschrittsbalken Demo
import tkinter as tk
from tkinter import ttk

class MarqueeProgressDemo:
    def __init__(self, root):
        self.root = root
        self.root.title("Marquee Fortschrittsbalken Demo")

        self.style = ttk.Style()
        self.style.layout("TMHorizontal.TProgressbar", [("Horizontal.Progressbar.trough", {"children": [("Horizontal.Progressbar.pbar", {"side": "left", "sticky": "ns"})], "sticky": "nswe"})])
        self.style.configure("TMHorizontal.TProgressbar", troughcolor="#E0E0E0", background="#00BFFF", lightcolor="#00BFFF", darkcolor="#00BFFF")

        self.progressbar = ttk.Progressbar(self.root, style="TMHorizontal.TProgressbar", orient='horizontal', mode='indeterminate')
        self.progressbar.pack(pady=10)

        self.progressbar.start()

if __name__ == "__main__":
    root = tk.Tk()
    app = MarqueeProgressDemo(root)
    root.mainloop()
```

```python
# TextProgressDemo
import tkinter as tk
from tkinter import ttk
import time

class TextProgressDemo:
    def __init__(self, root):
        self.root = root
        self.root.title("Textbasierte Fortschrittsanzeige Demo")

        self.progress_label = tk.Label(self.root, text="")
        self.progress_label.pack(pady=20)

        self.start_progress()

    def start_progress(self):
        for i in range(101):
            self.progress_label.config(text=f"Fortschritt: {i}%")
            self.root.update_idletasks()
            time.sleep(0.05)  # Simulieren von Fortschritt

if __name__ == "__main__":
    root = tk.Tk()
    app = TextProgressDemo(root)
    root.mainloop()
```

```python
# console_progress_bar
import time

# Funktion für eine textbasierte Fortschrittsanzeige in der Konsole
def console_progress_bar():
    total = 100
    for i in range(total + 1):
        progress = i / total * 100
        print(f"Ladevorgang: [{i * '#'}{(total - i) * '-'}] {progress:.1f}%", end='\r')
        time.sleep(0.1)

    print("\nFertig!")

if __name__ == "__main__":
    console_progress_bar()
```

```python
# mit threads 
import tkinter as tk
from tkinter import filedialog, messagebox
import pandas as pd
from threading import Thread

class ExcelMergerApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Excel Merger")

        # Frame für die Benutzeroberfläche
        self.frame = tk.Frame(self.root, padx=10, pady=10)
        self.frame.pack()

        # Button zum Hinzufügen von Excel-Dateien
        self.add_files_button = tk.Button(self.frame, text="Excel-Dateien hinzufügen", command=self.add_files)
        self.add_files_button.grid(row=0, column=0, padx=5, pady=5)

        # Dropdown-Menüs für Dateien, Arbeitsblätter und Spalten
        self.file_dropdown = tk.StringVar()
        self.file_dropdown.set("Datei auswählen")
        self.file_menu = tk.OptionMenu(self.frame, self.file_dropdown, "Datei auswählen")
        self.file_menu.grid(row=0, column=1, padx=5, pady=5)

        self.sheet_dropdown = tk.StringVar()
        self.sheet_dropdown.set("Arbeitsblatt auswählen")
        self.sheet_menu = tk.OptionMenu(self.frame, self.sheet_dropdown, "Arbeitsblatt auswählen")
        self.sheet_menu.grid(row=0, column=2, padx=5, pady=5)

        self.column_dropdown = tk.StringVar()
        self.column_dropdown.set("Spalte auswählen")
        self.column_menu = tk.OptionMenu(self.frame, self.column_dropdown, "Spalte auswählen")
        self.column_menu.grid(row=0, column=3, padx=5, pady=5)

        # Button zum Hinzufügen der ausgewählten Spalte
        self.add_column_button = tk.Button(self.frame, text="Spalte hinzufügen", command=self.add_column)
        self.add_column_button.grid(row=0, column=4, padx=5, pady=5)

        # Button zum Löschen der ausgewählten Spalte
        self.delete_column_button = tk.Button(self.frame, text="Spalte löschen", command=self.delete_column)
        self.delete_column_button.grid(row=0, column=5, padx=5, pady=5)

        # Button zum Zusammenführen und Speichern
        self.merge_button = tk.Button(self.frame, text="Daten zusammenführen und speichern", command=self.merge_and_save)
        self.merge_button.grid(row=1, column=0, columnspan=6, padx=5, pady=5)

        # Liste für ausgewählte Dateien und Spalten
        self.selected_files = []
        self.selected_columns = []  # List of tuples (file, sheet_name, column_name, column_index)

        # Neuer DataFrame für die ausgewählten Daten
        self.selected_data = pd.DataFrame()

        # Label für die angezeigten Spalten
        self.columns_label = tk.Label(self.frame, text="Ausgewählte Spalten:")
        self.columns_label.grid(row=2, column=0, columnspan=6, padx=5, pady=5)

        # Textfeld für angezeigte Spalten
        self.columns_text = tk.Text(self.frame, height=5, width=50)
        self.columns_text.grid(row=3, column=0, columnspan=6, padx=5, pady=5)

    def add_files(self):
        files = filedialog.askopenfilenames(filetypes=[("Excel Files", "*.xlsx;*.xls")])
        self.selected_files.extend(files)
        self.update_file_dropdown()

    def update_file_dropdown(self):
        self.file_menu['menu'].delete(0, 'end')
        self.file_dropdown.set("Datei auswählen")
        for file in self.selected_files:
            self.file_menu['menu'].add_command(label=file, command=lambda f=file: self.update_sheet_dropdown(f))

    def update_sheet_dropdown(self, file):
        self.sheet_menu['menu'].delete(0, 'end')
        self.sheet_dropdown.set("Arbeitsblatt auswählen")
        try:
            xls = pd.ExcelFile(file)
            for sheet_name in xls.sheet_names:
                self.sheet_menu['menu'].add_command(label=sheet_name, command=lambda s=sheet_name: self.update_column_dropdown(file, s))
        except Exception as e:
            messagebox.showerror("Fehler beim Einlesen der Datei", f"Fehler beim Einlesen von {file}: {str(e)}")

    def update_column_dropdown(self, file, sheet_name):
        self.column_menu['menu'].delete(0, 'end')
        self.column_dropdown.set("Spalte auswählen")
        try:
            df = pd.read_excel(file, sheet_name=sheet_name, nrows=0)  # Lese nur die Spaltenüberschriften
            for i, column in enumerate(df.columns):
                self.column_menu['menu'].add_command(label=column, command=lambda c=column, idx=i: self.select_column(c, file, sheet_name, idx))
        except Exception as e:
            messagebox.showerror("Fehler beim Einlesen der Spalten", f"Fehler beim Einlesen der Spalten von {file}, {sheet_name}: {str(e)}")

    def select_column(self, column, file, sheet_name, idx):
        self.column_dropdown.set(column)
        selected_info = (file, sheet_name, column, idx)
        if selected_info not in self.selected_columns:
            self.selected_columns.append(selected_info)
            self.update_selected_columns()
            # Füge die Spalte dem neuen DataFrame hinzu (im Hintergrund laden)
            t = Thread(target=self.load_column_data, args=(column, file, sheet_name, idx))
            t.start()

    def load_column_data(self, column, file, sheet_name, idx):
        try:
            df = pd.read_excel(file, sheet_name=sheet_name, usecols=[idx])
            self.selected_data[column] = df[column]
        except Exception as e:
            messagebox.showerror("Fehler beim Laden der Spalte", f"Fehler beim Laden der Spalte {column} von {file}, {sheet_name}: {str(e)}")

    def add_column(self):
        selected_column = self.column_dropdown.get()
        if selected_column != "Spalte auswählen":
            for file, sheet_name, column, idx in self.selected_columns:
                if column == selected_column:
                    messagebox.showwarning("Spalte bereits ausgewählt", f"Die Spalte '{selected_column}' wurde bereits hinzugefügt.")
                    return

            self.selected_columns.append((None, None, selected_column, None))  # Placeholder
            self.update_selected_columns()

    def delete_column(self):
        selected_column = self.column_dropdown.get()
        if selected_column != "Spalte auswählen":
            for info in self.selected_columns[:]:  # Iteriere über Kopie der Liste
                if info[2] == selected_column:
                    self.selected_columns.remove(info)
                    del self.selected_data[selected_column]
                    self.update_selected_columns()
                    messagebox.showinfo("Spalte gelöscht", f"Spalte '{selected_column}' wurde gelöscht.")
                    return
            messagebox.showwarning("Spalte nicht gefunden", "Bitte wählen Sie eine Spalte aus der Liste.")

    def update_selected_columns(self):
        self.columns_text.delete('1.0', tk.END)
        for file, sheet_name, column, idx in self.selected_columns:
            self.columns_text.insert(tk.END, f"{column}\n")

    def merge_and_save(self):
        if self.selected_data.empty:
            messagebox.showwarning("Keine Daten ausgewählt", "Bitte wählen Sie mindestens eine Spalte aus.")
            return

        # Speichern des zusammengeführten DataFrames in eine neue Excel-Datei (im Hintergrund speichern)
        t = Thread(target=self.save_data, args=())
        t.start()

    def save_data(self):
        save_path = filedialog.asksaveasfilename(defaultextension=".xlsx", filetypes=[("Excel Files", "*.xlsx")])
        if save_path:
            try:
                self.selected_data.to_excel(save_path, index=False)
                messagebox.showinfo("Speichern erfolgreich", f"Daten erfolgreich in {save_path} gespeichert.")
            except Exception as e:
                messagebox.showerror("Fehler beim Speichern", f"Fehler beim Speichern der Datei: {str(e)}")

def main():
    root = tk.Tk()
    app = ExcelMergerApp(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
